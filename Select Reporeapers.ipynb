```python
from ggplot import *
from pyspark.sql.types import *
from pyspark.sql.functions import *

import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = sqlContext.read.csv("dataset.csv", 
                         sep=",", 
                         header=True, 
                         inferSchema=True).cache()
aux = df.toPandas()
```

```python
aux['stars'] = aux['stars'].astype(int)
aux = aux.loc[(aux['size'] > 100000) &
        (aux['continuous_integration'] == 1) &
        (aux['scorebased_org'] == 1) &
        (aux['scorebased_utl'] == 1) &
        (aux['randomforest_org'] == 1) & 
        (aux['randomforest_utl'] == 1) &
        (aux['stars'].astype(int) > 100) &
        (aux['unit_test'] > 0.1)]
aux
```

```python
dfAux = sqlContext.createDataFrame(aux)
```
